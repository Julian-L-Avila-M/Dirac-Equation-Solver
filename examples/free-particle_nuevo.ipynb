# **Importación de libreria dirac_solver y adicionales:**

```python
import sys
sys.path.append('..')
import numpy as np
from dirac_solver.geometry import Grid
from dirac_solver.initial_state import InitialState, ConstantSpinor, GaussianPacket, PlaneWave
from dirac_solver.core import DiracProblemBuilder, DiracSolver
from dirac_solver import _core
```

---
# **Definición de Geometrias (mallas):**

```python
shape = (200,)
spacing = (0.1,)
grid_1D = Grid(shape, spacing)
```

```python
shape = (100,100)
spacing = (0.1,0.1)
grid_2D = Grid(shape, spacing)
```

```python
shape = (30,30,30)
spacing = (0.1,0.1,0.1)
grid_3D = Grid(shape, spacing)
```

---
# **Definición de estado inicial:**

## 1 dimensión:

```python
constant_spinor = ConstantSpinor(momentum=[5, 0, 0])
```

```python
initial_state_1D_GP = GaussianPacket(constant_spinor, center=[1, 0, 0], spatial_width=1.0)
```

```python
initial_state_1D_PW = PlaneWave(constant_spinor)
```

## 2 dimensiones:

```python
constant_spinor = ConstantSpinor(momentum=[5, 5, 0])
```

```python
initial_state_2D_GP = GaussianPacket(constant_spinor, center=[1, 1, 0], spatial_width=1.0)
```

```python
initial_state_2D_PW = PlaneWave(constant_spinor)
```

## 3 dimensiones:

```python
constant_spinor = ConstantSpinor(momentum=[5, 5, 5])
```

```python
initial_state_3D_GP = GaussianPacket(constant_spinor, center=[1, 1, 1], spatial_width=1.0)
```

```python
initial_state_3D_PW = PlaneWave(constant_spinor)
```

---
# **Definición de potenciales:**

```python
potential = _core.FreeParticle()
```

---
# **Definición de condiciones de frontera:**

```python
boundary_PB = _core.PeriodicBoundary()
```

```python
boundary_AB = _core.AbsorbingBoundary(strength=0.5, width=30)
```

---
# **Construcción de la simulación:**

```python
problem = DiracProblemBuilder() \
    .set_grid(grid_1D) \
    .set_initial_state(initial_state_1D_GP) \
    .set_potential(potential) \
    .set_boundary_condition(boundary_PB) \
    .set_time_parameters(time_step=0.001, total_time=5.0) \
    .build()

solver_1D_GP_PB = DiracSolver(problem)
```

# **Ejecución de la simulación:**

```python
solver_1D_GP_PB.run_simulation()
solver_1D_GP_AB.run_simulation()
solver_1D_PW_PB.run_simulation()
solver_1D_PW_AB.run_simulation()
solver_2D_GP_PB.run_simulation()
solver_2D_GP_AB.run_simulation()
solver_2D_PW_PB.run_simulation()
solver_2D_PW_AB.run_simulation()
```
