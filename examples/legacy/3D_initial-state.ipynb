# 3D Initial State Example

```python
import sys
sys.path.append('..')
import numpy as np
from dirac_solver.geometry import Grid
from dirac_solver.initial_state import GaussianPacket, ConstantSpinor
```

## 1. Create a Constant Spinor

```python
constant_spinor = ConstantSpinor(spin_polarization=[0, 0, 1], spin_superposition=[1 + 0j, 0 + 0j], momentum=[1, 0, 0], mass=0.511)
```

## 2. Create a Gaussian Packet Initial State

```python
initial_state = GaussianPacket(constant_spinor=constant_spinor, center=[0, 0, 0], spatial_width=1.5)
```

## 3. Evaluate the State at a Single Point

```python
position = [10, 10, 0]
psi_at_x = initial_state.evaluate(grid_position=position)
print(psi_at_x)
```

## 4. Evaluate the State on a 3D Grid

```python
grid3d = Grid(shape=(10, 10, 10), spacing=(0.5, 0.5, 0.5))
psi_on_grid = initial_state.evaluate_on_grid(grid3d)
print(psi_on_grid.shape)
```
