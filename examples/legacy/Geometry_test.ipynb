# Pruebas de la clase Grid (geometry.py)

Este notebook prueba todas las funcionalidades de la clase `Grid` de la librería **dirac_solver**:

1. Creación de mallas en 1D, 2D y 3D.
2. Verificación de atributos: `shape`, `dim`, `spacing`, `origin`.
3. Validación de coordenadas generadas (`coords`).
4. Cálculo del elemento de volumen `dv`.
5. Estimación de la condición de Courant `courant_limit()`.
6. Visualización de las mallas con `matplotlib`.


```python
# === Imports ===
import numpy as np
import matplotlib.pyplot as plt
from dirac_solver.geometry import Grid

# Configuración de plots
plt.style.use("seaborn-v0_8")
```

```python
# === Prueba 1: Grid 1D ===
grid1d = Grid(shape=(10,), spacing=0.1)

print("Grid 1D")
print("Shape:", grid1d.shape)
print("Dim:", grid1d.dim)
print("Origin:", grid1d.origin)
print("Coords shape:", grid1d.coords.shape)
print("dv:", grid1d.dv)
print("Courant limit:", grid1d.courant_limit())

# Graficar
x = grid1d.coords[:, 0]
plt.figure(figsize=(6,2))
plt.plot(x, np.zeros_like(x), "o-", label="Puntos de la malla")
plt.title("Grid 1D")
plt.xlabel("x")
plt.legend()
plt.show()
```

```python
# === Prueba 2: Grid 2D ===
grid2d = Grid(shape=(10, 8), spacing=(0.2, 0.3))

print("Grid 2D")
print("Shape:", grid2d.shape)
print("Dim:", grid2d.dim)
print("Origin:", grid2d.origin)
print("Coords shape:", grid2d.coords.shape)
print("dv:", grid2d.dv)
print("Courant limit:", grid2d.courant_limit())

# Graficar
x = grid2d.coords[:, :, 0].flatten()
y = grid2d.coords[:, :, 1].flatten()
plt.figure(figsize=(5,5))
plt.scatter(x, y, c="blue", s=20)
plt.title("Grid 2D")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.show()
```

```python
# === Prueba 3: Grid 3D ===
grid3d = Grid(shape=(6, 6, 6), spacing=(0.5, 0.5, 0.5))

print("Grid 3D")
print("Shape:", grid3d.shape)
print("Dim:", grid3d.dim)
print("Origin:", grid3d.origin)
print("Coords shape:", grid3d.coords.shape)
print("dv:", grid3d.dv)
print("Courant limit:", grid3d.courant_limit())

# Graficar (subsample para no saturar)
x = grid3d.coords[:, :, :, 0].flatten()
y = grid3d.coords[:, :, :, 1].flatten()
z = grid3d.coords[:, :, :, 2].flatten()

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x, y, z, c="red", s=10)
ax.set_title("Grid 3D")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```

```python
# === Prueba 4: Comparaciones de dv ===
# Con la modificacíon de constants

grid_a = Grid(shape=(10,), spacing=0.1)
grid_b = Grid(shape=(10,), spacing=0.2)

print("dv con spacing=0.1:", grid_a.dv)
print("dv con spacing=0.2:", grid_b.dv)
print("Relación esperada (dv_b / dv_a):", grid_b.dv / grid_a.dv)
```

```python
# === Prueba interna de la condición de Courant ===
# con la modificacíon de constants
import math as math

# Configuración de tolerancias
RTOL = 1e-12
ATOL = 1e-12

# Grid simple en 1D
grid_test = Grid(shape=(11,), spacing=0.1)
c = grid_test.c

# Valor calculado por la clase
dt_calc = grid_test.courant_limit()

# Valor esperado directo según la fórmula
inv_sq = sum((1.0/dx)**2 for dx in grid_test.spacing)
dt_expected = 1.0 / (c * math.sqrt(inv_sq))

# Mostrar resultados
print("=== Prueba interna Courant 1D ===")
print("dt (Grid.courant_limit) =", dt_calc)
print("dt (esperado)           =", dt_expected)

# Validar
assert np.isclose(dt_calc, dt_expected, rtol=RTOL, atol=ATOL), \
       f"Courant test falló: {dt_calc} vs {dt_expected}"

print("✅ Prueba Courant interna completada correctamente.")
```
