```python
#Prueba de Gridd 

# --- Prueba 1D ---
grid_1d = ds.Grid(shape=(11,), spacing=0.1)
print("Grid 1D coords:\n", grid_1d.coords)
print("Courant limit 1D:", grid_1d.courant_limit())

# --- Prueba 2D ---
grid_2d = ds.Grid(shape=(5,5), spacing=0.2)
print("Grid 2D coords shape:", grid_2d.coords.shape)
print("Courant limit 2D:", grid_2d.courant_limit())

# --- Prueba 3D ---
grid_3d = ds.Grid(shape=(3,3,3), spacing=0.3)
print("Grid 3D coords shape:", grid_3d.coords.shape)
print("Courant limit 3D:", grid_3d.courant_limit())
```

```python
#Prueba del spinnor 

spinor = ds.ConstantSpinor(spin_polarization=[0,0,1], spin_superposition=[1,0], momentum=[1,0,0])
print("Spinor constante 4x1:\n", spinor.constant_spinor)
```

```python
#prueba del paquete gaussiano

gauss = ds.GaussianPacket(constant_spinor=spinor, center=[0,0,0], spatial_width=1.0)
# Evaluar en algunos puntos
points = [[0,0,0], [0.5,0.5,0.5], [-0.5,-0.5,-0.5]]
for p in points:
    psi = gauss.evaluate(p)
    print(f"ψ({p}) = {psi.flatten()}")
```
