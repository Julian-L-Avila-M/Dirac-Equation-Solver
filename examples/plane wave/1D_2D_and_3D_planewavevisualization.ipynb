# 1D, 2D and 3D Simulation and Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from dirac_solver.geometry import Grid
from dirac_solver.initial_state import GaussianPacket, PlaneWave
from dirac_solver.core import DiracSolver, DiracProblemBuilder
from dirac_solver.storage import HDF5Storage
from dirac_solver import _core, electron_mass
from dirac_solver.initial_state import ConstantSpinor
from dirac_solver.storage import create_animation
```

```python
constant_spinor = ConstantSpinor(momentum=[5,0,0])
```

## 1D Simulation with periodic and absorbing  Boundary and each potential 

```python

# ==========================
# 1. Grid 1D
# ==========================
grid_1d = Grid(
    shape=(400,),         # 400 puntos
    spacing=(0.05,)       # dx = 0.05
)

# ==========================
# 2. Estado inicial
# ==========================
spinor = ConstantSpinor(momentum=[5.0, 0.0, 0.0])   # el p2 y p3 se ignoran
initial_state_1d = PlaneWave(constant_spinor)

# ==========================
# 3. Potencial 1D
# ==========================
def harmonic_1d(r):
    x = r[0]
    return 0.5 * x*x         # V(x) = x²/2

potential_1d = _core.ScalarPotential(harmonic_1d)

# ==========================
# 4. Frontera
# ==========================
boundary_1d = _core.AbsorbingBoundary(strength=0.5)

# ==========================
# 5. Construir el problema
# ==========================
builder_1d = DiracProblemBuilder()
problem_1d = (
    builder_1d
    .set_grid(grid_1d)
    .set_initial_state(initial_state_1d)
    .set_potential(potential_1d)
    .set_boundary_condition(boundary_1d)
    .set_time_parameters(time_step=0.01, total_time=10.0)
    .build()
)

# ==========================
# 6. Almacenamiento HDF5
# ==========================
storage_1d = HDF5Storage(
    filepath="1DAbsorbingscalar.hdf5",
    problem=problem_1d
)

# ==========================
# 7. Ejecutar la simulación
# ==========================
solver_1d = DiracSolver(problem_1d)
solver_1d.run_simulation(storage_handler=storage_1d, save_every_n_steps=5)

# Cerrar archivo (por seguridad)
storage_1d.close()
create_animation('1DAbsorbingscalar.hdf5', '1DAbsorbingscalar.gif')
```

## 2D Simulation with periodic and absorbing  Boundary and each potential 

```python
grid_2d = Grid(shape=(100, 100), spacing=(0.1, 0.1))
constant_spinor_2d = ConstantSpinor(momentum=[10.0, 30.0, 0.0])
initial_state_2d = PlaneWave(constant_spinor)

def harmonic_2d(r):
    x, y = r
    return 0.5 * (x*x + y*y)

potential_2d = _core.ScalarPotential(harmonic_2d)

boundary_2d = _core.AbsorbingBoundary(strength=0.5)
builder_2d = DiracProblemBuilder()
problem_2d = (builder_2d
              .set_grid(grid_2d)
              .set_initial_state(initial_state_2d)
              .set_potential(potential_2d)
              .set_boundary_condition(boundary_2d)
              .set_time_parameters(time_step=0.01, total_time=10.0)
              .build())
storage_2d = HDF5Storage(filepath='2DAbsorbingscalar.hdf5', problem=problem_2d)
solver_2d = DiracSolver(problem_2d)
solver_2d.run_simulation(storage_handler=storage_2d, save_every_n_steps=10)
create_animation('2DAbsorbingscalar.hdf5', '2DAbsorbingscalar.gif')
```
