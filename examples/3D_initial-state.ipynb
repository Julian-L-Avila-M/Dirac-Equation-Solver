```python
import dirac_solver
from dirac_solver.geometry import Grid
```

#### Se crea el Espinor Constante:

```python
constant_spinor = dirac_solver.ConstantSpinor(spin_polarization=[0,0,1], spin_superposition=[1 + 0j, 0 + 0j], momentum=[1,0,0], mass=0.511)
```

#### Se crea el estado (GaussianPacket):

```python
estado_inicial = dirac_solver.GaussianPacket(constant_spinor = constant_spinor, center = [0,0,0], spatial_width = 1.5)
```

#### Se evalua el estado en una posición exacta:

```python
posicion = [10, 10, 0]
psi_en_x = estado_inicial.evaluate(grid_position=posicion)
```

```python
print(psi_en_x)
```
